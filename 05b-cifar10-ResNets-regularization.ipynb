```python
BATCH_SIZE = 100
```

```python
train_sampler = SubsetRandomSampler(train_indicies)
train_loader = DataLoader(train_dataset,
                         BATCH_SIZE,
                         sampler=train_sampler)
val_sampler = SubsetRandomSampler(val_indicies)
val_loader = DataLoader(train_dataset,
                        BATCH_SIZE,
                        sampler=train_sampler)
```

```python
def accuracy(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    return torch.tensor(torch.sum(preds == labels).item() / len(preds))
```

```python
train_loader = ml.DeviceDataLoader(train_loader)
val_loader = ml.DeviceDataLoader(val_loader)
```

```python
val_loss, _, val_metric = ml.evaluate(model, F.cross_entropy, val_loader, metric=accuracy)
```

## Use entire training set for training, and use test set for validation

```python
# data transforms
stats = ((0.491, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
train_tfms = tt.Compose([tt.RandomCrop(32, padding=4, padding_mode="reflect"),
                        tt.RandomHorizontalFlip(),
                        tt.ToTensor(),
                        tt.Normalize(*stats)])

test_tfms = tt.Compose([tt.ToTensor(), tt.Normalize(*stats)])
```

```python
# Download the dataset
train_dataset = CIFAR10(root="./data",
                  download=True,
                  train=True,
                       transform=train_tfms)

test_dataset = CIFAR10(root="./data",
                  download=True,
                  train=False,
                      transform=test_tfms)
```

```python
BATCH_SIZE = 256
```

```python
train_loader = DataLoader(train_dataset,
                         BATCH_SIZE, shuffle=True,
                         num_workers=8, pin_memory=True)
val_loader = DataLoader(test_dataset,
                         BATCH_SIZE, shuffle=False,
                         num_workers=8, pin_memory=True)
```
