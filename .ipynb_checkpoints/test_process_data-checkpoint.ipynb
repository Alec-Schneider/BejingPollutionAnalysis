```python
# plot of the resulted trajectory        
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(20,20))
#axes[0].plot(Time, X.data, "k-", label="Actual", markersize=20)
ax1.plot(imputed_pm25.index, dingling[pm25].values, "g-", label="Filtered position", markersize=1)
ax2.plot(imputed_pm25.index, prep.filtered_state_means[:, 0] + prep._sigma, "r--", label="+ sigma", markersize=1)
ax2.plot(imputed_pm25.index, prep.filtered_state_means[:, 0] - prep._sigma, "r--", label="- sigma", markersize=1)
ax1.grid()
ax2.grid()
ax1.legend(loc="upper left")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.show()  
```
